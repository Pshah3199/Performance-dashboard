```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample operational performance data
data = {
    "Employee": ["A", "B", "C", "D", "E"],
    "Tasks_Completed": [12, 15, 8, 20, 14],
    "Delay_Days": [2, 0, 5, 1, 0]
}

df = pd.DataFrame(data)

print("Performance Data:")
print(df)

print("\nAverage Tasks Completed:", df["Tasks_Completed"].mean())
print("Total Delays (days):", df["Delay_Days"].sum())
```

```python
plt.figure()
plt.bar(df["Employee"], df["Tasks_Completed"])
plt.xlabel("Employee")
plt.ylabel("Tasks Completed")
plt.title("Tasks Completed by Employees")
plt.show()
```

```python
plt.figure()
plt.bar(df["Employee"], df["Delay_Days"])
plt.xlabel("Employee")
plt.ylabel("Delay (Days)")
plt.title("Task Delays by Employees")
plt.show()
```

## Performance Dashboard

This project analyzes operational performance data such as task completion and delays.
The dashboard helps identify productivity trends and areas requiring improvement.

### Insights
- Employees with higher task completion show better efficiency
- Delay data helps identify workload or process issues
